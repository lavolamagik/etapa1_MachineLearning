## 2. Resumen del artículo base (Clifford et al., 2017)

El PhysioNet/CinC Challenge 2017 fue un concurso para desarrollar algoritmos que clasifiquen el ritmo cardíaco (Normal, Fibrilación Auricular (FA), Otro, Ruidoso) usando un único registro corto (30-60s) de ECG de una sola derivación.

La FA es la arritmia más común y su detección es crucial, pero desafiante debido a su naturaleza episódica y la necesidad de métodos robustos al ruido. El desafío utilizó un gran dataset (8,528 registros de entrenamiento) y evaluó los algoritmos con la métrica F1 promediada en un conjunto de prueba oculto.

Se promovió el desarrollo de algoritmos de código abierto y la presentación de los resultados en la conferencia Computing in Cardiology, destacando la importancia de las características derivadas de los intervalos RR para la clasificación.

## 4.Visualización de distribuciones por clase (Normal vs AFib) usando matplotlib o seaborn

```python
print("\n## Visualización de Distribuciones por Clase")

features = ['mean_rr', 'std_rr', 'skew_rr', 'kurt_rr']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data=df, x=feature, hue='label', kde=True, palette={'Normal': 'green', 'AFib': 'red'})
    plt.title(f'Distribución de {feature} por Clase')
    plt.xlabel(feature)
    plt.ylabel('Frecuencia')
    plt.legend(title='Clase')
plt.tight_layout()
plt.show()

# También podemos usar boxplots para una mejor visualización de la mediana, cuartiles y outliers
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(data=df, x='label', y=feature, palette={'Normal': 'lightgreen', 'AFib': 'salmon'})
    plt.title(f'Boxplot de {feature} por Clase')
    plt.xlabel('Clase')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```

5.1 ¿Qué variables parecen diferenciar más entre clases?

De acuerdo con los histogramas y boxplots analizados, las variables que mejor permiten diferenciar entre las clases Normal y AFib son principalmente std_rr y mean_rr(este en menor medida que std_rr). Estas variables muestran diferencias en sus distribuciones y medianas según la clase, lo que puede ser útil para clasificación o diagnóstico. Por otro lado, skew_rr presenta poca capacidad de discriminación, mientras que kurt_rr muestra superposición entre clases, lo que muestra que seria muy limitado usarlo para difernciar clases. viendo esto y analisandolo, podemos concluir que: los intervalos RR (std_rr) es el factor más distintivo entre pacientes normales y aquellos con fibrilación auricular.
